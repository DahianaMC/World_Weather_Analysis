```python
# Import the dependencies.
import pandas as pd
import gmaps
import requests
# Import the API key.
from config import g_key
```

```python
# Upload the CSV you saved and was created in Weather_Database.
city_data_df = pd.read_csv("weather_data/WeatherPy_challenge.csv")
city_data_df.head()
```

```python
city_data_df.dtypes
```

```python
# Prompt the customer to add a minimum and maximum temperature value.
min_temp = float(input("What is the minimum temperature you would like for your trip? "))
max_temp = float(input("What is the maximum temperature you would like for your trip? "))
```

```python
# Filter the data for temperature inputs.
city_data_temp = city_data_df.loc[(city_data_df["Max Temp"] >= min_temp) & (city_data_df["Max Temp"] <= max_temp)]
city_data_temp
```

```python
# Prompt the customer if they would like to be raining or not
city_raining = str(input("Do you want it to be raining? (yes/no)"))
# Prompt the customer if they would like to be snowing or not
city_snowing = str(input("Do you want it to be snowing? (yes/no)"))
```

```python
# Creating dataframe with the yes or no inputs from raining and snowing
if (city_raining.lower() == "yes" and city_snowing.lower() == "yes" ):
    rain_snow_df = city_data_temp.loc[(city_data_temp["Rain inches (last 3 hrs)"] != 0)
                                      & (city_data_temp["Snow inches (last 3 hrs)"] != 0)]
    if (rain_snow_df.empty == True):
        print("Has not been raining or snowing for the last 3 hours on the cities")

elif (city_raining.lower() == "yes" and city_snowing.lower() == "no" ):
    rain_snow_df = city_data_temp.loc[(city_data_temp["Rain inches (last 3 hrs)"] != 0) 
                                      & (city_data_temp["Snow inches (last 3 hrs)"] == 0)]
    print("Has not been raining for the last 3 hours on the cities")
    
elif (city_raining.lower() == "no" and city_snowing.lower() == "yes" ):
    rain_snow_df = city_data_temp.loc[(city_data_temp["Rain inches (last 3 hrs)"] == 0) 
                                      & (city_data_temp["Snow inches (last 3 hrs)"] != 0)]
    print("Has not been snowing for the last 3 hours on the cities")

elif (city_raining.lower() == "no" and city_snowing.lower() == "no" ):
    rain_snow_df = city_data_temp.loc[(city_data_temp["Rain inches (last 3 hrs)"] == 0)
                                      & (city_data_temp["Snow inches (last 3 hrs)"] == 0)]

# In case someone has a typo error    
else:
     print("Please enter yes or no... Ask the question again")
        
rain_snow_df
```

```python
# Add the cities to a marker layer map with a pop-up marker for each city that includes
if (rain_snow_df.empty == True):
    print("No data to report")
else:
    hotel_df = rain_snow_df[["City", "Country", "Max Temp", "Current Description", "Lat", "Lng"]].copy()
    hotel_df["Hotel Name"] = ""
    print(hotel_df.head(10))
```

```python
# Get Travel Destinations
# Set parameters to search for a hotel.
params = {
    "radius": 5000,
    "type": "lodging",
    "key": g_key
}
```

```python
# Iterate Through hotel_df DataFrame
for index, row in hotel_df.iterrows():
    # Get the latitude and longitude.
    lat = row["Lat"]
    lng = row["Lng"]

    # Add the latitude and longitude to location key for the params dictionary.
    params["location"] = f"{lat},{lng}"

    # Use the search term: "lodging" and our latitude and longitude.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    # Make request and get the JSON data from the search.
    hotels = requests.get(base_url, params=params).json()
    # Grab the first hotel from the results and store the name.
    try:
        hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    except (IndexError):
        print("Hotel not found... skipping.")

hotel_df.head(10)
```

```python
# Remove rows with no names for hotel names
hotel_df_final = hotel_df[hotel_df["Hotel Name"] != '']
hotel_df_final
```

```python
# Save and upload the new DataFrame as WeatherPy_vacation.csv
output_data_file = "weather_data/WeatherPy_vacation.csv"
# Export the hotel data into a CSV.
hotel_df_final.to_csv(output_data_file, index_label = "City ID")
```

```python
# Adding a pop-up marker including: Hotel Name, City, Country and weather description
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Description</dt><dd>{Current Description} and {Max Temp} °F</dd>
</dl>
"""
```

```python
# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in hotel_df_final.iterrows()]
```

```python
gmaps.configure(api_key=g_key)
from ipywidgets.embed import embed_minimal_html
# Create a marker layer map.
locations = hotel_df_final[["Lat", "Lng"]]
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(marker_layer)

# Save the figure.
embed_minimal_html('weather_data/WeatherPy_vacation_map.html', views=[fig])

# Call the figure to plot the data.
fig

```
